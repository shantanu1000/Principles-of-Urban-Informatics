```python
## Task 4
url = 'https://raw.githubusercontent.com/nychealth/coronavirus-data/7953c97d1e58bbed9934ea04affb12ca74d9c0fb/data-by-modzcta.csv'

df_COVID = pd.read_csv(url)

df_COVID. head()
```

```python
df_AGG_COVID = df_COVID[['BOROUGH_GROUP','COVID_CASE_COUNT']].groupby(by = 'BOROUGH_GROUP').sum()
df_AGG_COVID
```

```python
df_AGG_COVID.plot.bar()
```

```python
df_percentage_COVIDCASES = df_COVID[['BOROUGH_GROUP','COVID_CASE_COUNT','TOTAL_COVID_TESTS']].groupby(by = 'BOROUGH_GROUP').sum()
df_percentage_COVIDCASES['%_of_POSITIVE_CASES']= (100*(df_percentage_COVIDCASES['COVID_CASE_COUNT']/df_percentage_COVIDCASES['TOTAL_COVID_TESTS'])).round(2)
df_percentage_COVIDCASES
```

```python
df_percentage_COVIDCASES['%_of_POSITIVE_CASES'].plot.bar(ylabel='%_of_POSITIVE_CASES' )
```

```python
df_AGG_COVID_POP = df_COVID[['BOROUGH_GROUP','POP_DENOMINATOR','COVID_CASE_COUNT']].groupby(by = 'BOROUGH_GROUP').sum()
df_AGG_COVID_POP
```

```python
df_AGG_COVID_POP['%_of_POP_WITH_COVID']= (100*(df_AGG_COVID_POP['COVID_CASE_COUNT']/df_AGG_COVID_POP['POP_DENOMINATOR'])).round(2)
df_AGG_COVID_POP
                                              
```

```python
df_AGG_COVID_POP['%_of_POP_WITH_COVID'].plot.bar(ylabel='%_of_POP_WITH_COVID')
```
